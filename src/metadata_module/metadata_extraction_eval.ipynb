## 2. Run Analysis on One Image

## 3. Batch Test a Folder

```python
batch_results = analyse_folder("../../data/sample_images")

for result in batch_results:
    if "error" in result:
        print(f"\nError in {result['file']}: {result['error']}")
        continue
        
    print()
    print(f"File: {result['file']}")
    print("score: ", result['ai_probability'])
    print("decision: ", result['decision'])
    print("rationale: ")
    
    for item in result['rationale']:
        print(f"\t - {item}")
```
